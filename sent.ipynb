```python
import urllib.request, json, xmltodict
import numpy as np
import pandas as pd
import re


def get_revs_for_app(appID):
    response = urllib.request.urlopen("https://itunes.apple.com/ru/rss/customerreviews/id=%s/sortBy=mostRecent/xml" % appID)
    j = xmltodict.parse(response.read())
    data = json.dumps(j)
    response.close()
    d = json.loads(data)

    feed = d.get("feed")
    entrys = feed.get("entry")
    reviews = []
    index = 0

    for entry in entrys:

        rating = entry.get("im:rating")
        if rating == None:
            continue

        title = entry.get("title")
        content = entry.get("content")[0].get("#text")
        author = entry.get("author").get("name")

        if rating != None:
            index = index + 1
            # print str(index) + ": " + rating + " --- " + author + " --- " + title + " --- " + content
            review = [rating, author, title, content]
            reviews.append(review)
    return reviews

```

```python
revs = pd.DataFrame(reviews, columns=["rating", "author", "title", "content"])
revs.drop('author', inplace=True, axis=1)
revs.drop('title', inplace=True, axis=1)
revs.to_csv("res.csv")

revs

my_own_dict = ["–¥–æ–Ω–∞—Ç", "–≤—ã–ª–µ—Ç–∞—Ç—å", "—Ç–æ—Ä–º–æ–∑–∏—Ç—å", "–¥–æ–±–∞–≤–∏—Ç—å"]

def bad_words_count(line):
    count = 0
    for word in my_own_dict:
        count += sum(1 for _ in re.finditer(r'%s' % re.escape(word), line))
    return count

revs['bad_words'] = revs['content'].apply(bad_words_count)
revs
```

```python
#drop eng reviews
revs.drop(revs[revs.content.str.contains ("[A-Za-z]")].index, inplace=True)
#drop reviews with rating 3-4
revs.drop(revs[(revs.rating == "3") | (revs.rating == "4")].index, inplace=True)

revs
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def strtonormalform(line, should_delete = {'PREP', 'CONJ', 'PRCL', 'INTJ'}):
    ans = ''
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        lex = lexInfo.normal_form
        if (lexInfo.tag.POS not in should_delete) and (re.match(r"[-+]?\d+$", lex) is None) :#—Å–æ—é–∑—ã, –ø—Ä–µ–¥–ª–æ–≥–∏ –∏ —Ç.–ø. –Ω–µ —Å–æ—Ö—Ä–∞–Ω—è–µ–º
            ans = ans + ' ' + lex
    return ans

revs['content'] = revs['content'].apply(strtonormalform)
revs.to_pickle('revs')
revs

#revs['bad_words'] = [1 if for text in revs.content]
```

```python
import sklearn.metrics

def trainTestClf(clf, X_train, y_train, X_test):
    clf.fit(X_train, y_train)
    return clf.predict(X_test)

def printF2(y_test, y_pred, vType ='tf-idf', cType='Bayess'):
    print("Vectorizing: ", vType, ". Classifier: ", cType, sep='')

    print("accuracy = %.3f" % sklearn.metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
    print("precision = %.3f" % sklearn.metrics.precision_score(y_true = y_test, y_pred = y_pred))
    print("recall = %.3f" % sklearn.metrics.recall_score(y_true = y_test, y_pred = y_pred))
    print("F-–º–µ—Ä–∞ = %.3f" % sklearn.metrics.f1_score(y_true = y_test, y_pred = y_pred))
```
