```python
import urllib.request, json, xmltodict
import numpy as np
import pandas as pd
import re


def get_revs_for_app(appID):
    reviews = []
    for i in range(1, 11):
        revs_url = "https://itunes.apple.com/ru/rss/customerreviews/id={0}/page={1}/sortBy=mostRecent/xml".format(appID, i)
        print ("fetching:  ", revs_url, '\n')
        response = urllib.request.urlopen(revs_url)
        j = xmltodict.parse(response.read())
        data = json.dumps(j)
        response.close()
        d = json.loads(data)

        feed = d.get("feed")
        entrys = feed.get("entry")
        
        index = 0
        
        if entrys != None:
            for entry in entrys:
                if (type(entry) == str):
                    continue
                
                rating = entry.get("im:rating")
                if rating == None:
                    continue

                title = entry.get("title")
                content = entry.get("content")[0].get("#text")
                author = entry.get("author").get("name")

                if rating != None:
                    index = index + 1
                    # print str(index) + ": " + rating + " --- " + author + " --- " + title + " --- " + content
                    review = [rating, author, title, content]
                    reviews.append(review)
    return reviews

```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def strtonormalform(line, should_delete = {'PREP', 'CONJ', 'PRCL', 'INTJ'}):
    ans = ''
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        lex = lexInfo.normal_form
        if (lexInfo.tag.POS not in should_delete) and (re.match(r"[-+]?\d+$", lex) is None) :#союзы, предлоги и т.п. не сохраняем
            ans = ans + ' ' + lex
    return ans

revs['content'] = revs['content'].apply(strtonormalform)
#revs.to_pickle('revs')
#revs.to_csv("clean_reviews.csv")
revs

#revs['bad_words'] = [1 if for text in revs.content]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.decomposition import TruncatedSVD, NMF
from sklearn.model_selection import RandomizedSearchCV
from nltk.corpus import stopwords
import string
```

```python
import sklearn.metrics

def trainTestClf(clf, X_train, y_train, X_test):
    clf.fit(X_train, y_train)
    return clf.predict(X_test)

def printF2(y_test, y_pred, vType ='tf-idf', cType='Bayess'):
    print("Vectorizing: ", vType, ". Classifier: ", cType, sep='')

    print("accuracy = %.3f" % sklearn.metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
    print("precision = %.3f" % sklearn.metrics.precision_score(y_true = y_test, y_pred = y_pred))
    print("recall = %.3f" % sklearn.metrics.recall_score(y_true = y_test, y_pred = y_pred))
    print("F-мера = %.3f" % sklearn.metrics.f1_score(y_true = y_test, y_pred = y_pred))
```

```python
vectorizers = [TfidfVectorizer, CountVectorizer]
classifiers = [LogisticRegression, LinearSVC]

for c in classifiers:
    for v in vectorizers:
        model = Pipeline([("vectorizer", v()), ("classifier", c())])
        print(str(c().__class__.__name__) + ' ' + str(v().__class__.__name__))
        print(cross_val_score(model, revs.content, revs.rating, scoring='accuracy', cv = 5).mean())
        print
```

```python
# +биграммы

for c in classifiers:
    for v in vectorizers:
        model = Pipeline([("vectorizer", v(ngram_range=(1, 2))), ("classifier", c())])
        print(str(c().__class__.__name__) + ' ' + str(v().__class__.__name__))
        print(cross_val_score(model, revs.content, revs.rating, scoring='accuracy', cv = 5).mean())
        print
```

```python
# + стоп-слова

for c in classifiers:
    for v in vectorizers:
        model = Pipeline([("vectorizer", v(ngram_range=(1, 2), stop_words=stopwords.words('russian'))), ("classifier", c())])
        print(str(c().__class__.__name__) + ' ' + str(v().__class__.__name__))
        print(cross_val_score(model, revs.content, revs.rating, scoring='accuracy', cv = 5).mean())
        print
```

```python
# + буквенные n-граммы

for c in classifiers:
    for v in vectorizers:
        model = Pipeline([("vectorizer", v(analyzer='char_wb', ngram_range=(3, 5), stop_words=stopwords.words('russian'))), ("classifier", c())])
        print(str(c().__class__.__name__) + ' ' + str(v().__class__.__name__))
        print(cross_val_score(model, revs.content, revs.rating, scoring='accuracy', cv = 5).mean())
        print
```
