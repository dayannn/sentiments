```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
def strtonormalform(line, should_delete = {'PREP', 'CONJ', 'PRCL', 'INTJ'}):
    ans = ''
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        lex = lexInfo.normal_form
        if (lexInfo.tag.POS not in should_delete) and (re.match(r"[-+]?\d+$", lex) is None) :#союзы, предлоги и т.п. не сохраняем
            ans = ans + ' ' + lex
    return ans

revs['content'] = revs['content'].apply(strtonormalform)
#drop empty reviews
revs.drop(revs[revs.content == ""].index, inplace=True)
#revs.to_csv("clean_reviews.csv")
revs
```

```python
tone = np.zeros((len(norm_texts), 4))
```

```python
# общая тональность по словарю

for i, text in enumerate(norm_texts_splitted):
    count = 0
    for j, word in enumerate(text):        
        s = Dict.get(word, 0)
        if s != 0:
            count += 1
            if j >= 1:
                if text[j-1][0] == 'не':
                    s = - s
            tone[i, 0] += s
    if count:
        tone[i, 0] = tone[i, 0]/count
```

```python
# количество восклицательных знаков

for i, text in enumerate(initial_revs.content):
    for e in text:
        if e == '!':
            tone[i, 1] += 1
```

```python
# количество глаголов в повелительном наклонении

initial_revs.drop(initial_revs.index.difference(revs.index), inplace=True)
initial_revs['rating'] = revs['rating']

def getImperatives(line, should_delete = {'PREP', 'CONJ', 'PRCL', 'INTJ'}):
    ans = 0
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        if (lexInfo.tag.POS == 'VERB'):
            if (lexInfo.tag.mood == 'impr'):
                ans += 1                
    return ans

initial_revs['imps'] = [getImperatives(line) for line in initial_revs.content]
```

```python
tone[:,2] = list(initial_revs['imps'])
```

```python
interrogative_words = ["где", "как", "зачем", "почему", "какой"]

for i, text in enumerate(norm_texts_splitted):
    count = 0
    for j, word in enumerate(text):        
        if (word in interrogative_words):
            count += 1
    tone[i, 3] = count
```

```python
import sklearn.metrics

def trainTestClf(clf, X_train, y_train, X_test):
    clf.fit(X_train, y_train)
    return clf.predict(X_test)

def printF2(y_test, y_pred, vType ='tf-idf', cType='Bayess'):
    print("Vectorizing: ", vType, ". Classifier: ", cType, sep='')

    print("accuracy = %.3f" % sklearn.metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
    print("precision = %.3f" % sklearn.metrics.precision_score(y_true = y_test, y_pred = y_pred))
    print("recall = %.3f" % sklearn.metrics.recall_score(y_true = y_test, y_pred = y_pred))
    print("F-мера = %.3f" % sklearn.metrics.f1_score(y_true = y_test, y_pred = y_pred))
```

```python
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
clf = svm.LinearSVC()
clf2 = MultinomialNB()
clf3 = RandomForestClassifier()
```

```python
# без доп признаков
X_train, X_test, y_train, y_test = train_test_split(cvect, initial_revs.rating, test_size=0.1, random_state=42)

printF2(y_test, trainTestClf(clf, X_train, y_train, X_test), vType = "Boolean", cType = 'svm')
print('\n')
#printF2(y_test, trainTestClf(clf2, X_train, y_train, X_test), vType = "Boolean", cType = 'Naive Bayes')
print('\n')
printF2(y_test, trainTestClf(clf3, X_train, y_train, X_test), vType = "Boolean", cType = 'Random Forest')
```

```python
# с доп признаками
X_train, X_test, y_train, y_test = train_test_split(train, initial_revs.rating, test_size=0.1, random_state=42)

printF2(y_test, trainTestClf(clf, X_train, y_train, X_test), vType = "Boolean", cType = 'svm')
print('\n')
#printF2(y_test, trainTestClf(clf2, X_train, y_train, X_test), vType = "Boolean", cType = 'Naive Bayes')
print('\n')
printF2(y_test, trainTestClf(clf3, X_train, y_train, X_test), vType = "Boolean", cType = 'Random Forest')
```
