```python
sent = pd.read_csv('words_sent.csv', sep=';', header=None, names =['word', 'sent', '1', '2', '3'])
Dict = sent.groupby('word')['sent'].mean().to_dict()

# нормальзованное значение тональности
def get_dict_sent(line):
    count = 0
    res = 0
    prev_word = ''
    for j, word in enumerate(line.split()):
        neg = False
        if (word == "нет" or word == "нету"):
            res += -0.25
        if word.startswith("не_"):
            s = Dict.get(word[3:], 0)
            neg = True
        else:
            s = Dict.get(word, 0)

        if s != 0:
            if (prev_word in ["крайне", "очень", "много", "сильно", "большой"]):
                s *= 1.6
            if (prev_word in ["чуть", "немного" "несильно"]):
                s *= 0.5
            count += 1
            if (neg):
                res += -s
            else:
                res += s
        if (s == 0 and neg):
            res += -0.25
            
        prev_word = word
    if count:
        res = res/count
        
    return res

def get_words_num(line):
    count = 0
    for word in line.split():
        count += 1
    return count


def get_excl_marks_num(line):
    num = 0
    for e in line:
        if e == '!':
            num += 1
            # tone[i, 1] = 1
    return num


def getImperatives(line):
    ans = 0
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        if lexInfo.tag.POS == 'VERB':
            if lexInfo.tag.mood == 'impr':
                ans += 1

    return ans


interrogative_words = ["где", "как", "зачем", "почему", "какой"]

def get_interrog_words(line):
    count = 0
    for j, word in enumerate(line.split()):
        if word in interrogative_words:
            count += 1

    for e in line:
        if e == '?':
            count += 1

    return count


def get_digits_num(line):
    count = 0
    for letter in line:
        if letter >= "0" and letter <= "9":
            count += 1
    return count


def find_elongated_words(line):
    if (line == ""):
        return False

    prev_letter = line[0]
    flag = 0
    count = 0
    for letter in line[1:]:
        if letter == prev_letter:
            count += 1
        else:
            count = 0;
            prev_letter = letter
        if count >= 2:
            flag = 1
    return flag

def count_capital_letters(line):
    count = 0
    for letter in line:
        if letter >= "А" and letter <= "Я":
            count += 1
    return count

def getPastTenseVerbs(line, should_delete = {'PREP', 'CONJ', 'PRCL', 'INTJ'}):
    ans = 0
    words = re.findall(r"[\w]+[\w|-]*", line)
    for word in words:
        lexInfo = morph.parse(word)[0]
        if (lexInfo.tag.POS == 'VERB'):
            if (lexInfo.tag.tense == 'past'):
                ans += 1
    if (ans > 0):
        return ans
    else:
        return 0
    return ans

def get_sad_smiles_num(line):
    num = 0
    for e in line:
        if e == '(':
            num += 1
    if (':с' in line):
        num += 1
    return num

def get_adversatives_num(line):
    sum = 0
    words = re.findall(r"[\w]+[\w|-]*", line)
    words = [item.lower() for item in words]
    for word in words:
        if word in ["даже", "но", "а", "только",  "если", "раньше", "теперь"]:
            sum += 1
        
    return sum
```
